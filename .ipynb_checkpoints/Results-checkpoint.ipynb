# Анализ результатов

```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.image import imread
from IPython.display import Image, display
```

Загрузим результаты по 5 моделям:

```python
scores = pd.read_csv('data/scores.csv')
scores
```

Отобразим результаты на графике:

```python
labels = scores['model']
score = scores['score']
time = scores['time']

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots()
ax2 = ax.twinx()
rects1 = ax.bar(x - width/2, score, width, label='score', color='lightsteelblue')
rects2 = ax2.bar(x + width/10, time, width/5, label='time', color='darkorange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_title('Scores for all models (100000 samples)')
ax.set_xticks(x - width/2.5)
ax.set_xticklabels(labels)
ax.set_ylabel('AUC')
ax.set_yticks(np.arange(0.7, 0.8, 0.01))
ax.set_ylim([0.7, 0.8])
ax.legend(handles=[rects1, rects2])
ax2.set_ylabel('Time, sec')
ax2.set_yticks(np.arange(0, 750, 50))
ax2.set_ylim([0, 700])

for rect in rects1:
    height = rect.get_height()
    ax.annotate('{}'.format(f'{height:0.3f}'), 
                xy=(rect.get_x() + rect.get_width() / 2, height), 
                xytext=(0, 2), textcoords='offset points', 
                ha='center', va='bottom')
    
for rect in rects2:
    height = rect.get_height()
    ax2.annotate('{}'.format(f'{height:3.0f}'), 
                xy=(rect.get_x() + rect.get_width() / 2, height), 
                xytext=(0, 2), textcoords='offset points', 
                ha='center', va='bottom')

fig.tight_layout()
plt.show();
```

**Видим, что модель LightGBM (далее LGBM) и ансамблевая модель из нейросетей дают наилучшие результаты метрики.** 

При этом модель LGBM быстрее в 4 раза, чем нейросетевая модель. Сравним матрицы ошибок двух наилучших моделей:

```python
for filename in os.listdir('pictures/'):
    if filename.startswith('LightGBM'):
        display(Image(filename=os.path.join('pictures', filename), width=400, height=400))
```

Для модели LGBM без учета весов характерно такое поведение, при котором она очень редко блокирует отрицательные кейсы, и довольно часто пропускает положительные кейсы. Это связано с тем, что модель впринципе плохо предсказывает положительный класс (16 меток на 100000 примеров). Такая модель в продакшене будет являться несостоятельной.

Модель LGBM с весами дает немного меньшее значение целевой метрики, зато умеет размечать положительные классы. Можно сказать, что на каждого пропущенного мошенника модель точно определит двух мошенников. При этом модель довольно часто блокирует отрицательные кейсы.

**Теперь рассмотрим ансамблевую модель:**

```python
ensemble_list = ['Ensemble_model_3.png', 'Ensemble_model_4.png', 'Ensemble_model_5.png', 
                 'Ensemble_model_8.png', 'Ensemble_model_9.png', 'Ensemble_model_10.png']

for filename in ensemble_list:
    display(Image(filename=os.path.join('pictures', filename), width=400, height=400))
```

Здесь на под ансамблевой моделью понимается модель класса EnsembleNet, написанная в рамках этой работы. Модели, входящие в ансамбль, обучаются на отдельных чанках каждая, а результат предсказания выбирается по среднему или методом голосования. Сразу можем отметить, что особой разницы между усреднением и голосованием нет, но с усреднением результаты лучше.

Так же как и с моделью LGBM, если не учитывать веса то модель теряет возможность предсказывать положительный класс. Поэтому в функционале модели EnsembleNet предусмотрен учет весов, и мы наблюдаем хорошие результаты по матрице ошибок: такая модель в среднем на каждого пропущенного мошенника верно определяет 3 мошенников (то есть recall выше, чем у LGBM). Так же верно определено большее число положительных классов: 2376 против 2135 у LGBM. Минусом является большее время обучения модели.
