# LightGBM model

Модель LightGBM обладает впечатляющими характеристиками по скорости обучения и точности предсказаний. LightGBM расширяет алгоритм градиентного бустинга, добавляя тип автоматического выбора объектов, а также фокусируясь на примерах бустинга с большими градиентами. Это может привести к резкому ускорению обучения и улучшению прогнозных показателей. 

```python
import os
import gc
import itertools
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb

from sklearn.metrics import RocCurveDisplay, roc_curve, auc, classification_report, confusion_matrix
from sklearn.utils import class_weight
from datetime import datetime
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
 
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
 
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
 
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
start = datetime.now()
```

Прочитаем подготовленный датасет:

```python
path_to_dataset = 'data/processed_data/train'

dataset_paths = sorted([os.path.join(path_to_dataset, filename) for filename in os.listdir(path_to_dataset) 
                        if filename.startswith('processed_chunk')])

frames = []

for step, chunk_path in enumerate(dataset_paths):
    print(f'{step}. chunk_path {chunk_path}')
    frames.append(pd.read_parquet(chunk_path, columns=None))

df = pd.concat(frames, axis=0).fillna(0.)
X, y = df.drop(['id', 'flag'], axis=1), df['flag']
train_data = lgb.Dataset(X, label=y)

data = pd.read_parquet('data/processed_data/test/test.parquet').head(100000)
validation_data = lgb.Dataset(data.drop(['id', 'flag'], axis=1), label=data['flag'], reference=train_data)

del df, frames
gc.collect()
```

## Обучение без весов

Выберем оптимальные параметры, результаты метрик будем записывать в scores:

```python
scores = []

param = {'num_leaves': 31, 'objective': 'binary', 'metric': 'auc'}

num_round = 1000
bst = lgb.train(param, train_data, num_round, valid_sets=[validation_data])
```

```python
X_test = data.drop(['id', 'flag'], axis=1)
y_test = data['flag']
```

```python
y_pred = np.round(bst.predict(X_test))
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, y_pred), classes=['0','1'],
                      title='Confusion matrix | LightGBM | No weights')
plt.savefig('pictures/LightGBM_model_1.png', bbox_inches='tight');
```

**Без учета весов модель помечает только 16 объектов из 100000 как положительный класс. Значение метрики recall практически равняется нулю.**

```python
# Посчитаем значение метрики AUC и построим ROC кривую
y_pred_proba = bst.predict(X_test)
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
scores.append(roc_auc)

RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc).plot()
plt.title('LightGBM | No weights')
plt.savefig('pictures/LightGBM_model_2.png', bbox_inches='tight');
```

**Значение метрики auc выше целевого показателя в 0.75, но из-за того что она не умеет предсказывать положительный класс, такую модель нельзя использовать в продакшене.**

## Обучение с весами

Рассчитаем веса и разметим вектор весов классов для каждого объекта:

```python
param = {'num_leaves': 31, 'objective': 'binary', 'metric': 'auc'}

class_weights = class_weight.compute_class_weight(class_weight='balanced', 
                                                  classes=np.unique(y), y=y)
weights = np.zeros_like(y).astype(np.float32)
weights[y == 0] = class_weights[0]
weights[y == 1] = class_weights[1]
train_data.set_weight(weights)

num_round = 1000
bst = lgb.train(param, train_data, num_round, valid_sets=[validation_data])
```

```python
X_test = data.drop(['id', 'flag'], axis=1)
y_test = data['flag']
```

```python
y_pred = np.round(bst.predict(X_test))
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(confusion_matrix(y_test, y_pred), classes=['0','1'],
                      title='Confusion matrix | LightGBM | With weights')
plt.savefig('pictures/LightGBM_model_3.png', bbox_inches='tight');
```

**Здесь наблюдаем иную картину: модель научилась предсказывать положительный класс. В среднем на одного пропущенного мошенника модель верно определяет 2 мошенников. При этом довольно много кейсов ошибочно помечаются как положительные, что не так критично в банкинге.**

```python
# Посчитаем значение метрики AUC и построим ROC кривую
y_pred_proba = bst.predict(X_test)
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
scores.append(roc_auc)

RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc).plot()
plt.title('LightGBM | With weights')
plt.savefig('pictures/LightGBM_model_4.png', bbox_inches='tight');
```

**Метрика auc немного уменьшилась, но по вышеописанным причинам мы не можем использовать модель без весов.**

# Результаты

Сохраним лучшие результаты в таблицу scores:

```python
df_scores = pd.read_csv('data/scores.csv')
df_scores.loc[df_scores['model'] == 'LightGBM', 'score'] = max(scores)

time = datetime.now() - start
df_scores.loc[df_scores['model'] == 'LightGBM', 'time'] = int(time.seconds)
df_scores.head()
```

```python
df_scores.to_csv('data/scores.csv', index=False)
```
